## At a glance

Key Terms:

1) Sequential processing.
2) Parallel processing.
3) SPMD paradigm.
  - True parallelism.
  - Virtual parallelism (concurrency).
4) CUDA built-in variable.
5) CUDA memory row-major Layout

---

   - 1) Sequential processing reaches a peak performance - Sequential processing reaches a peak in performance due to constraints generated by the increasing clock frequency, which leads to elevated energy consumption and high heat dissipation.
   - 2) Parallel processing -< Independent operations >- We can parallelize independent operations.
   - 3) SPMD paradigm - a single program being executed by multiple processing elements, each working on different pieces of data.
       - `True parallelism` (parallelism): This execution can be parallel if the system has multiple processors or cores, allowing truly simultaneous execution.
       - `Virtual parallelism` (concurrency): SPMD approach doesn't strictly require multiple physical cores to work. Even on a single-core system, the tasks can be scheduled in such a way that each gets a slice of time to execute. The operating system switches between these tasks rapidly (this is called context switching), creating the appearance that they're all running at the same time, even though only one task is actually running at any given moment. This is what we refer to as concurrent execution.
       - > Saying "concurrently" is more general and accurate because it encompasses both true parallel execution and situations where the execution is interleaved but still logically overlapping.
   - 4) CUDA built-in variables - Since we are launching multiple kernel instances (or thread) on different subsets of data, it's essential to track the specific data each thread operates on. CUDA provides built-in variablessuch as `threadIdx`, `blockIdx`, `blockDim`, and `gridDim` to uniquely identify each thread and its context within the grid and blocks. 
  
  

    - > `nvcc` compile the code to PTX (Parallel Thread Execution) a low-level Virtual Machine.
   - 5) CUDA row-major layout. There are other ways to store an n-dimensional array in a one-dimensional memory space (e.g., column-major layout).
